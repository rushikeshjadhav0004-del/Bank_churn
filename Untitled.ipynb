# Correlation Analysis

```python
x=df.drop(['Geography','Gender','Exited','Surname','CustomerId'],axis=1)
y=df['Exited']
```

```python
plt.figure(figsize=(8,7))
sns.heatmap(x.corr(),annot=True);
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report, accuracy_score
```

```python
num_col=x.columns
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
sm = SMOTE(random_state=42)
x_train_sm, y_train_sm = sm.fit_resample(x_train, y_train)
```

```python
preprocessor = ColumnTransformer([('num', StandardScaler(),num_col)
                                
])
```

```python
pipe_lr = Pipeline([
    ('preprocessing', preprocessor),
    ('model', LogisticRegression(class_weight='balanced'))
])
```

```python
pipe_rf = Pipeline([
    ('preprocessing', preprocessor),
    ('model', RandomForestClassifier())
])
```

```python
pipe_svm = Pipeline([
    ('preprocessing', preprocessor),
    ('model', SVC())
])
```

```python
models = {
    "Logistic Regression": pipe_lr,
    "Random Forest": pipe_rf,
    "SVM": pipe_svm,
}

for name, model in models.items():
    print(f"\n {name}")
    
    model.fit(x_train_sm, y_train_sm)
    y_pred = model.predict(x_test)
    
    print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("Train_accuracy",pipe_rf.score(x_train_sm,y_train_sm))
print("Test_accuracy",pipe_rf.score(x_test,y_test))
```

# Cross_validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, x, y, cv=5)
print(scores.mean())
```

# Hyper_Parameter_Tunning

```python
param_grid = {
    'model__n_estimators': [100, 200],
    'model__max_depth': [10, None],
    'model__min_samples_split': [2, 10],
    'model__min_samples_leaf': [1, 5],
    'model__max_features': ['sqrt', 'log2']
}
```

```python
grid_search = GridSearchCV(
    estimator=pipe_rf,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)
grid_search.fit(x_train_sm, y_train_sm)
best_model = grid_search.best_estimator_
```

```python
print("Best Parameters:", grid_search.best_params_)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = best_model.predict(x_test)

print("Train Accuracy:", best_model.score(x_train, y_train))
print("Test Accuracy:", best_model.score(x_test, y_test))

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
y_prob = best_model.predict_proba(x_test)[:,1]

y_pred_new = (y_prob > 0.3).astype(int)

print(classification_report(y_test, y_pred_new))
```

# Boosting_Technique

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
pipe_gb = Pipeline([
    ('preprocessing', preprocessor),
    ('model', GradientBoostingClassifier(random_state=42))
])
```

```python
param_grid = {
    'model__n_estimators': [100, 200],
    'model__learning_rate': [0.05, 0.1, 0.2],
    'model__max_depth': [3, 5],
    'model__min_samples_split': [2, 10],
    'model__min_samples_leaf': [1, 5]
}
```

```python
grid_search = GridSearchCV(
    estimator=pipe_gb,
    param_grid=param_grid,
    cv=5,
    scoring='f1',  
    n_jobs=-1,
    verbose=2
)
grid_search.fit(x_train,y_train)
```

```python
print("Best Params:", grid_search.best_params_)
best_model = grid_search.best_estimator_
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = best_model.predict(x_test)

print("Train Accuracy:", best_model.score(x_train, y_train))
print("Test Accuracy:", best_model.score(x_test, y_test))

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
